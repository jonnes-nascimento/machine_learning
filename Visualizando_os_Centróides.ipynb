# Gerando os dados para o teste

```
from sklearn.datasets import make_blobs
import pandas as pd

# gera um conjunto de 1000 registros para teste
dados, _ = make_blobs(random_state=7, n_samples=1000, n_features=2)

# transforma o conjunto de dados em um dataframe do pandas
dados = pd.DataFrame(dados, columns=['coluna_1', 'coluna_2'])
dados.head()
```

# Plotando o gráfico

```
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# plota o grafico base com os dados criados pelo make_blob
plt.figure(figsize=(10,10))
plt.scatter(x=dados.coluna_1, y=dados.coluna_2)

# cria o modelo KMeans com 3 grupos
modelo = KMeans(n_clusters=3)
grupos = modelo.fit_predict(dados)

# plota um novo grafico, identificando os grupos e alterando as cores
plt.scatter(x=dados.coluna_1, y=dados.coluna_2, c=grupos, cmap='viridis')

# geramos os centroides
centroides = modelo.cluster_centers_

# plotamos o grafico novamente, com a marcacao dos centroides
plt.scatter(centroides[:, 0], centroides[:,1], marker='X', s=200, linewidths=5, color='r', zorder=8)
```
